```python
import numpy as np
import matplotlib.pyplot as plt
srate = 100

start_time = -2
end_time = 2

time = np.arange(start_time, end_time, 1/srate)

time[0:10]
```

```python
def create_signal(x):
    y = np.sin(2 * np.pi * x)
    return y
signal = create_signal(time)

plt.plot(time,signal)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
w0 = 10*np.pi
f0 = w0/(2*np.pi)
T0 = 1/f0

k = 2

fs = k*f0
Ts = 1/fs

srate = fs
start_time = -2*T0
end_time = 2*T0

continuous_time = np.arange(start_time,end_time, 1/(srate*1000))
continuous_signal = np.cos(10*np.pi*continuous_time)

time = np.arange(start_time,end_time, 1/srate)
signal = np.cos(10*np.pi*time)

plt.plot(continuous_time,continuous_signal)
plt.plot(time,signal, 'ro')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
