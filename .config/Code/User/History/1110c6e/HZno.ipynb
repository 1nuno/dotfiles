```python
import requests as r
from lxml import etree
from time import sleep
from bs4 import BeautifulSoup as bs
```

```python
def get_pessoas_json(dom,soup):
    pessoas_json = {}
    pessoas_nomes = dom.xpath('//div[@class = "search-results__list"]//h4[@class = "search-results__article-person-title"]/text()')
    pessoas_dados = soup.select('div.search-results__article-wrapper')
    for pessoa,nome in zip(pessoas_dados,pessoas_nomes):
        dom2 = etree.HTML(str(pessoa))
        info = [i for i in dom2.xpath('//li[@class = "search-results__details-list-item"]//text()') if i not in '\n,\n\n,\n\r,\r\n,\r'.split(',')]
        for i in range(0,len(info),2):
            if i != len(info)-1:
                if nome in pessoas_json:
                    pessoas_json[nome][info[i]] = info[i+1]
                else:
                    pessoas_json[nome] = {}
                    pessoas_json[nome][info[i]] = info[i+1]
    return pessoas_json
```
